```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:\\squid_games.csv")
df.head()
```

```python
df = df.drop(columns="user_location" , axis = 1)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns="user_description" , axis = 1)
df.head()
```

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
df["text"] = df["text"].apply(clean)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud , STOPWORDS

text = " ".join(i for i in df.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["text"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["text"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["text"]]
data = df[["text", "Positive", "Negative", "Neutral"]]
print(data.head())
```

```python
file_path = "E:\\squid_games_analysis.xlsx"

# Export the 'data' dataframe to an Excel file
data.to_excel(file_path, index=False)
```

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a > b) and (a > c):
        return "Positive 😊"
    elif (b > a) and (b > c):
        return "Negative 😠"
    else:
        return "Neutral 🙂"

result = sentiment_score(x, y, z)
print("Overall Sentiment:", result)
```
